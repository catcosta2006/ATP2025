c) mergeListas: defina uma função que recebe duas listas de números ordenadas crescentemente como
argumento e devolve uma nova lista ordenada de forma crescente com os elementos das duas listas.

```python
lista1 = [1, 7, 3, 6, 5, 9]
lista2 = [8, 4, 2]

def mergeListas(lista1, lista2):
    listaA = sorted(lista1)
    listaB = sorted(lista2)
    if listaA == []:   
        return listaB
    if listaB == []:
        return listaA
    if listaA[0] <= listaB[0]:
        return [listaA[0]] + mergeListas(listaA[1:], listaB)
    else:
        return [listaB[0]] + mergeListas(listaA, listaB[1:])

print(mergeListas(lista1, lista2))
```

a) splitLista: defina uma função que recebe dois argumentos, uma lista de números e um número, e dá
como resultado um tuplo composto por duas listas, a primeira contendo os números menores ou iguais ao
número passado como argumento, e a segunda contendo os números maiores que o argumento.

```python
lista1 = [1,2,3,4,5]

def splitLista(lista, n):
    res = []
    res1 = []
    for elem in lista:
        if elem <= n:
            res.append(elem)
        elif elem > n:
            res1.append(elem)
    return (res, res1)
    

print(splitLista(lista1, 3))    
```

b) multiplos23: defina uma função que calcula quantos múltiplos de 23 existem no intervalo de 1 a n em que n é o argumento da função

```python
def multiplos_23(n):
    res = []
    for elem in range(1, n, 23):
        if elem % 23 == 0:
            res.append(elem)
    return len(res)

print(multiplos_23(100))
```

d) existe_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome de um médico, e verifica se o médico existe nessa unidade hospitalar, devolvendo True caso exista e False em caso contrário
